```python
# prompt: view all slices of nii.gz file

import nibabel as nib
import matplotlib.pyplot as plt

# Load the NIfTI file
img = nib.load('/kidney_instance-1_annotation-1.nii.gz')  # Replace 'your_file.nii.gz' with the actual filename

# Get the image data as a NumPy array
data = img.get_fdata()

# Get the dimensions of the image
x_dim, y_dim, z_dim = data.shape

# Iterate through slices and display them
for i in range(z_dim):
  plt.figure()
  plt.imshow(data[:, :, i], cmap='gray')
  plt.title(f'Slice {i}')
  plt.show()
```

```python
# prompt: view all slices of nii.gz file except those that have only black (no contours of organs)

import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np

# Load the NIfTI file
img = nib.load('/imaging.nii.gz')  # Replace 'your_file.nii.gz' with the actual filename

# Get the image data as a NumPy array
data = img.get_fdata()

# Get the dimensions of the image
z_dim, x_dim, y_dim = data.shape

# Iterate through slices and display only those with non-zero values
for i in range(x_dim):
  slice_data = data[i, :, :]
  if np.any(slice_data != 0):  # Check if the slice contains any non-zero values
    plt.figure()
    plt.imshow(slice_data, cmap='gray')
    plt.title(f'Slice {i}')
    plt.show()
```
